ssl tern sparsity

```python
def weight_tern(fp_w, t_factor):
    fp_w = torch.from_numpy(fp_w)
    mean_w = fp_w.abs().mean()
    max_w = fp_w.abs().max()
    th = t_factor*max_w
#     th = 1e-4
#     print('th:{}'.format(th))
    output = fp_w.clone().zero_()
    W = fp_w[fp_w.ge(th)+fp_w.le(-th)].abs().mean()
    output[fp_w.ge(th)] = W
    output[fp_w.lt(-th)] = -W
#     print('W:{}'.format(W))
    return output.numpy()


# checkpoint = torch.load('./save/2019-03-11/cifar10_tern_resnet20_120_tern2ssl_00015_nogradfffl/model_best.pth.tar')
# checkpoint = torch.load('./save/2019-03-13/imagenet_resnet18b_ff_lf_tex1_50_tern2ssl_ff_lf_000005_group64_contine14/model_best.pth.tar')
# checkpoint = torch.load('./save/2019-03-14/cifar10_tern_resnet20_200_ssl2tern_fffl_0002_adam/model_best.pth.tar')
# checkpoint = torch.load('./save/2019-03-14/cifar10_tern_resnet20_200_tern2ssl_fffl_0002_adam/checkpoint.pth.tar')
# checkpoint = torch.load('./save/2019-03-14/cifar10_resnet20_120_ssl_1_adam/model_best.pth.tar')
# checkpoint = torch.load('./save/2019-03-14/imagenet_resnet18b_ff_lf_tex1_44_tern2ssl_ff_lf_0000035_group64_continue6/model_best.pth.tar')
# checkpoint = torch.load('./save/2019-03-14/cifar10_tern_resnet20_200_sslandtern_adam/model_best.pth.tar')
checkpoint = torch.load('./save/2019-03-16/imagenet_resnet18b_ff_lf_tex1_50_ssl2tern_ff_lf_group64/model_best.pth.tar')

params = checkpoint['state_dict']
counter = 0
overall = 0
t_factor = 0.05
count_all_density = 0
count_all_data = 0
for k,v in params.items():
      if 'layer' in k and 'conv' in k:
#     if 'weight' in k and 'conv' in k or 'classifier.weight' in k:
#     if 'weight' in k:
        counter += 1
        conv_w = v
#         print(conv_w.shape)
        output = np.array(conv_w)
        output = weight_tern(output, t_factor) 
        print(output.size)
        density = 1.0*(np.count_nonzero(output))/output.size
        print('{} density: {:.8f}'.format(k, density))
        count_all_density += np.count_nonzero(output)
        count_all_data += output.size

print('overall density: {:.6f}'.format((count_all_density)/count_all_data))       
```

```python
def weight_tern(fp_w, t_factor):
    fp_w = torch.from_numpy(fp_w)
    mean_w = fp_w.abs().mean()

    max_w = fp_w.abs().max()
    th = t_factor*max_w
#     th = 1e-4
    output = fp_w.clone().zero_()
    W = fp_w[fp_w.ge(th)+fp_w.le(-th)].abs().mean()
    output[fp_w.ge(th)] = W
    output[fp_w.lt(-th)] = -W
    
    return output.numpy()

# checkpoint = torch.load('./save/2019-03-11/cifar10_tern_resnet20_200_tern2ssl_fffl_0002_group16_mask/model_best.pth.tar')
# checkpoint = torch.load('./save/2019-03-11/cifar10_tern_resnet20_200_ssl2tern_005_mask/model_best.pth.tar')
# checkpoint = torch.load('./save/2019-03-11/cifar10_tern_resnet20_120_tern2ssl_00015_nogradfffl/model_best.pth.tar')
# checkpoint = torch.load('./save/2019-03-12/imagenet_resnet18b_ff_lf_tex1_50_tern_ssl_ffffl_00002_group64/model_best.pth.tar')
# checkpoint = torch.load('./save/2019-03-13/imagenet_resnet18b_ff_lf_tex1_50_tern2ssl_ff_lf_000005_group64_contine14/checkpoint.pth.tar')
# checkpoint = torch.load('./save/2019-03-14/cifar10_tern_resnet20_200_ssl2tern_fffl_0002_adam/model_best.pth.tar')
# checkpoint = torch.load('./save/2019-03-14/cifar10_tern_resnet20_200_tern2ssl_fffl_0002_adam/checkpoint.pth.tar')
# checkpoint = torch.load('./save/2019-03-14/imagenet_resnet18b_ff_lf_tex1_44_tern2ssl_ff_lf_0000035_group64_continue6/model_best.pth.tar')

# checkpoint = torch.load('./save/2019-03-14/cifar10_tern_resnet20_200_sslandtern_adam/model_best.pth.tar')
checkpoint = torch.load('./save/2019-03-16/imagenet_resnet18b_ff_lf_tex1_50_ssl2tern_ff_lf_group64/model_best.pth.tar')

params = checkpoint['state_dict']

counter = 0
overall = 0
t_factor = 0.05
num_one = 0
all_num_one = 0.0
all_num_group = 0.0
num_g = 4
all_sparsity_mean = 0.0
all_sparsity_min = 1.0
for k,v in params.items():
    
    if 'layer' in k and 'conv' in k:
#     if 'weight' in k:

#         print(k)
      
        counter += 1
        conv_w = v 
        output = np.array(conv_w)
        cout = output.shape[0]
        cin = output.shape[1]
        kh = output.shape[2]
        kw = output.shape[3]
        output = weight_tern(output, t_factor)
#         print(output)

        w_t = output.reshape(cout, cin // num_g, num_g, 3, 3)
        num_group = (cout * cin) // num_g
        w_t = w_t.reshape(num_group, num_g, 3, 3)
        print(w_t.shape)
        all_num_group += num_group
        sparsity_chall = np.zeros(num_group)
        num_one = 0
        for i in range(num_group):       
            sparsity_group = 1.0*(w_t[i].size - np.count_nonzero(w_t[i]))/w_t[i].size
            sparsity_chall[i] = sparsity_group
#             print('group wise {} sparsity: {:.8f}'.format(i, sparsity_group))
            if int(sparsity_group) is 1:
                num_one += 1
        all_num_one += num_one 
        print('num of sparse group:{}'.format(num_one))

        
        sparsity_mean = np.mean(sparsity_chall)
        sparsity_min = np.min(sparsity_chall)
        all_sparsity_mean += sparsity_mean
        if sparsity_min < all_sparsity_min:
            all_sparsity_min = sparsity_min
        print('{} filter wise sparsity_mean: {:.6f}\t sparsity_min: {:.6f}'.format(k, sparsity_mean, sparsity_min))
print('group density for all layer:{:.6f}'.format(1 - all_num_one/all_num_group))
print('all_sparsity_mean:{:.6f}'.format(all_sparsity_mean/counter))
print('all_sparsity_min:{:.6f}'.format(all_sparsity_min))
```
