```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
 
```

```python
plt.rcParams['font.size'] = 20.
plt.rcParams['axes.labelsize'] = 20.
plt.rcParams['xtick.labelsize'] = 20.
plt.rcParams['ytick.labelsize'] = 20.
plt.figure(figsize=(10,5))
x_slim = [0.25, 0.5, 0.75, 1.0]
slim = [79.5,85.6, 88.4,89.8]
x_inv = [0.16, 0.4, 0.6, 1.0]
invidual = [75.36, 87.15, 91.89, 91.93]
x_ours = [0.25,0.4,0.6,1.0]
y_ours = [86.68, 87.3, 88.89, 91.89]
x_ours1 = [0.25,0.6, 1.0]
y_ours1 = [86.44, 90.62, 91.89]
plt.plot(x_slim,slim, marker='s', linestyle='-', color='r',label='slim')
plt.plot(x_inv,invidual, marker='s', linestyle='-', color='b',label='individual')
plt.plot(x_ours,y_ours, marker='s', linestyle='-', color='g',label='ours-2subnet')
plt.plot(x_ours1,y_ours1, marker='s', linestyle='-', color='black',label='ours-1subnet')
plt.axvline(x=0.6, linestyle='--', color='black')

plt.xlabel('Model Size');
plt.ylabel('Accuracy');  
plt.legend(loc='lower right')
plt.yticks(np.arange(65.0, 100.0, 5.0))
fig1 = plt.gcf()
plt.draw()

fig1.savefig('adptive_cifar10.jpg')
plt.show()
```
