```python
import sys
import torch
import numpy as np
```

```python
dir = 'logs/mobilenetv2_exp_decaying/loss_accuracy.txt'
```

```python
train1x = []
train075x = []
train05x = []
train035x = []
test1x = []
test075x = []
test05x = []
test035x = []
epoch = []
total = []
train = []
count = 0
with open(dir, 'r') as mytxt:
    for line in mytxt:
        if count == 0:
            print(line)
            count+=1
        else:
            #epoch
            epoch_txt = line.split()[0]
            epoch.append(epoch_txt)
            
            #test
            test1x_txt = line.split()[5]
#             print(test1x_txt)
            test1x.append(test1x_txt)            

            test075x_txt = line.split()[6]
            test075x.append(test075x_txt)   
            
            test05x_txt = line.split()[7]
            test05x.append(test05x_txt) 
            
            test035x_txt = line.split()[8]
            test035x.append(test035x_txt) 
            
            #train1x
epoch = np.array(epoch)
epoch = epoch.astype(np.float)

test1x = np.array(test1x)
test1x = test1x.astype(np.float)

test075x = np.array(test075x)
test075x = test075x.astype(np.float)

test05x = np.array(test05x)
test05x = test05x.astype(np.float)

test035x = np.array(test035x)
test035x = test035x.astype(np.float)
```

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  

fig = plt.figure(1, figsize=(6, 6))
plt.rcParams.update({'font.size': 13})
plt.plot(epoch, test1x, c='red', label='test1x')
plt.plot(epoch, test075x, c='blue', label='test075x')
plt.plot(epoch, test05x, c='black', label='test05x')
plt.plot(epoch, test035x, c='green', label='test05x')

# plt.plot(arr2_2[3:], arr2_3[3:],'--', c='red', label='w/ gradient correctness-train')
# plt.plot(arr3_2, arr3_3, c='blue', label='w/o gradient correctness-test')
# plt.plot(arr4_2, arr4_3,'--', c='blue', label='w/o gradient correctness-train')

plt.legend(loc='best')
# ax1.set_title('A')
plt.ylabel('Accuracy (%)')
plt.xlabel('Epoch')

plt.show()
```
